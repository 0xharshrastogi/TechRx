```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.tokenize import word_tokenize
import glob
import os
```

```python
nltk.download("stopwords")
nltk.download("wordnet")
```

```python
def cleanText(text):
    text = (text.translate(str.maketrans("", "", string.punctuation))).lower()
    text = re.sub(r"\d+", "", re.sub(r"\[\d+\]", "", text))
    return text
```

```python
cleanText("What are you doing these days. Anything special?")
```

```python
input_folder = r"fetched_data"
input_files = glob.glob(os.path.join(input_folder, "*.txt"))

clean_data = {}

for i in input_files:
    with open(i, "r", encoding="utf-8") as f:
        text = f.read()
        clean_data[os.path.splitext(os.path.basename(i))[0].split("_")[-1]] = cleanText(
            text
        )
        f.close()

print(clean_data)
```

```python
for i, j in clean_data.items():
    print(i)
    with open(f"clean_data/clean_text_{i}.txt", "w", encoding="utf-8") as f:
        f.write(j)
        f.close()
```
